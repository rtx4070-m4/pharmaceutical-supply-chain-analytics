# Pharmaceutical Supply Chain Data Analysis
This notebook analyzes pharmaceutical supply chain data to identify delivery delays, supplier performance, inventory patterns, and cost insights.

**Tools Used:** Python, Pandas, Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_excel('pharmaceutical_supply_chain_dataset.xlsx')
df.head()
```

## Dataset Information

```python
df.info()
```

## Summary Statistics

```python
df.describe()
```

## Missing Values

```python
df.isnull().sum()
```

## Orders by Product Type

```python
product_counts = df['Product_Type'].value_counts()
product_counts
```

## Plot: Product Demand

```python
product_counts.plot(kind='bar')
plt.title('Product Demand')
plt.xlabel('Product Type')
plt.ylabel('Number of Orders')
plt.show()
```

## Average Delivery Lead Time by Supplier

```python
supplier_leadtime = df.groupby('Supplier_Name')['Lead_Time_Days'].mean().sort_values()
supplier_leadtime
```

## Plot: Supplier Delivery Performance

```python
supplier_leadtime.plot(kind='bar')
plt.title('Average Delivery Time by Supplier')
plt.xlabel('Supplier')
plt.ylabel('Lead Time (Days)')
plt.show()
```

## Inventory Levels by Warehouse

```python
warehouse_inventory = df.groupby('Warehouse_Location')['Inventory_Level'].mean()
warehouse_inventory
```

## Plot: Inventory Distribution

```python
warehouse_inventory.plot(kind='bar')
plt.title('Average Inventory Level by Warehouse')
plt.xlabel('Warehouse')
plt.ylabel('Inventory Level')
plt.show()
```

## Transport Mode Usage

```python
transport_counts = df['Transport_Mode'].value_counts()
transport_counts
```

## Plot: Transportation Mode Distribution

```python
transport_counts.plot(kind='bar')
plt.title('Transport Mode Distribution')
plt.xlabel('Transport Mode')
plt.ylabel('Number of Shipments')
plt.show()
```

## Key Insights
- Identify suppliers with the fastest delivery times.
- Understand which pharmaceutical products are most demanded.
- Monitor inventory levels across warehouses.
- Analyze transportation modes used in pharmaceutical logistics.
